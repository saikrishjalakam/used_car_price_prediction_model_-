```python
import numpy as np
```

```python
web= wb.Chrome("chromedriver.exe")
from tqdm import tqdm

for link in tqdm(all_urls):
    try:
        web.get(link) 
        time.sleep(5)   
    except:
        ad_id.append(np.nan)
        brand.append(np.nan)        
        year.append(np.nan)
        kms_driven.append(np.nan)
        fuel.append(np.nan)
        model.append(np.nan)
        variant.append(np.nan)
        color.append(np.nan)
        price.append(np.nan)
        location.append(np.nan)
        #Facing.append(np.nan)
        #Project_Name.append(np.nan)
        #Listed_by.append(np.nan)
    try:
        Link.append(link) 
        price.append(web.find_element_by_class_name('hoLrKb').text)
        location.append(web.find_element_by_class_name('bJxiZx').text)
        details = web.find_elements_by_class_name('sc-1sbb775-1')
        d = dict()
        for i in details:
            d[i.text.split('\n')[0]] = i.text.split('\n')[1]
        
#         print(d)
        
        if 'Ad ID' in d.keys():
            ad_id.append(d['Ad ID'])
        else:
            ad_id.append(np.nan)

        if 'Brand Name' in d.keys():
            brand.append(d['Brand Name'])
        else:
             brand.append(np.nan)  

        if 'Year' in d.keys():
            year.append(d['Year'])
        else:
            year.append(np.nan)

        if 'Kms Driven' in d.keys():
            kms_driven.append(d['Kms Driven'])
        else:
            kms_driven.append(np.nan)

        if 'Fuel Type' in d.keys():
            fuel.append(d['Fuel Type'])
        else:
            fuel.append(np.nan)

        if 'Model' in d.keys():
            model.append(d['Model'])
        else:
            model.append(np.nan)

        if 'Variant' in d.keys():
            variant.append(d['Variant'])
        else:
            variant.append(np.nan)

        if 'Color' in d.keys():
            color.append(d['Color'])
        else:
            color.append(np.nan)

    
    except:
        ad_id.append(np.nan)
        brand.append(np.nan)        
        year.append(np.nan)
        kms_driven.append(np.nan)
        fuel.append(np.nan)
        model.append(np.nan)
        variant.append(np.nan)
        color.append(np.nan)
        price.append(np.nan)
        location.append(np.nan)
        #Facing.append(np.nan)
        #Project_Name.append(np.nan)
        #Listed_by.append(np.nan)
```
